```python
ingredients_dataset = [ [i for i in items if i not in ("sugar", "salt")] for items in ingredients_dataset]
len(ingredients_dataset)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

To save memory we represent the transaction data in the sparse format. This is especially useful when we have lots of products and small transactions.

```python
te = TransactionEncoder()
te_ary = te.fit(ingredients_dataset).transform(ingredients_dataset)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
print (rules)
```

```python
support=rules.as_matrix(columns=['support'])
confidence=rules.as_matrix(columns=['confidence'])
```

```python
import random
import matplotlib.pyplot as plt
```

```python
for i in range (len(support)):
   support[i] = support[i] + 0.0025 * (random.randint(1,10) - 5) 
   confidence[i] = confidence[i] + 0.0025 * (random.randint(1,10) - 5)
 
plt.scatter(support, confidence,   alpha=0.5, marker="*")
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
 
```

```python
import networkx as nx
fig, ax=plt.subplots(figsize=(10,6))
GA=nx.from_pandas_edgelist(rules,source='antecedents',target='consequents')
nx.draw(GA,with_labels=True)
plt.show()
```

Note: that the entries in the "itemsets" column are of type frozenset, which is built-in Python type that is similar to a Python set but immutable, which makes it more efficient for certain query or comparison operations. Since frozensets are sets, the item order does not matter and it shows the frequent set occuring of the ingredients.
